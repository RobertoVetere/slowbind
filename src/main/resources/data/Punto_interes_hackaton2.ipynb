```python
# Calcular la frecuencia de visitas por barrio
intensidad_counts = df['BARRI'].value_counts()

# Calcular percentiles
percentiles = pd.qcut(intensidad_counts, q=10, labels=False, duplicates='drop')

# Crear un diccionario de mapeo para asignar puntajes del 1 al 10 basados en los percentiles
rating_mapping = {barrio: int(1 + 9 * perc / 9) for barrio, perc in zip(intensidad_counts.index, percentiles)}

# Crear la nueva columna 'INTENSIDAD' en el DataFrame original
df['INTENSIDAD'] = df['BARRI'].map(rating_mapping)


```
